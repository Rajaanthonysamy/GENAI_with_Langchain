```python
from langchain_community.document_loaders import TextLoader

loader=TextLoader("/home/raja/learnings/GEN_AI_KRISH/Langchain/1-Langchain/DataInjection/speech.txt")
loader
docs=loader.load()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_docs=text_splitter.split_documents(docs)
final_docs
```

```python
## Vector embedding and Vector StoreDB
from langchain_community.vectorstores import Chroma
db=Chroma.from_documents(final_docs,embedding_1024)
```

```python
db.__dict__
```

```python
search_query="It will be all the easier for us to conduct ourselves as belligerents"
retrived_res=db.similarity_search(search_query)
```
